```python
# ============================================================
#  Decision-Aware Leukemic Blast Detection — كل شيء في خلية واحدة
# ============================================================
import os, re, json, math, random, copy
import numpy as np
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms, models
from sklearn.metrics import roc_auc_score, accuracy_score
from scipy.stats import mannwhitneyu
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch
from PIL import Image

# ----- reproducibility -----
SEED=42; random.seed(SEED); np.random.seed(SEED)
torch.manual_seed(SEED); torch.cuda.manual_seed_all(SEED)
torch.backends.cudnn.deterministic=True; torch.backends.cudnn.benchmark=False
DEVICE="cuda" if torch.cuda.is_available() else "cpu"

# ----- paths (عدّلها) -----
DATA_DIR="/content/drive/MyDrive/C-NMC_training_data"
OUT_DIR="/content/drive/MyDrive/hematology_project_outputs"
FIG_DIR=os.path.join(OUT_DIR,"figures")
os.makedirs(OUT_DIR,exist_ok=True); os.makedirs(FIG_DIR,exist_ok=True)

# ربط Drive + نسخ محلي للتسريع
try:
    from google.colab import drive
    if not os.path.isdir("/content/drive/MyDrive"): drive.mount("/content/drive")
    import shutil; LOCAL="/content/cnmc_local"
    if os.path.isdir(DATA_DIR) and not os.path.isdir(LOCAL):
        print("نسخ محلي..."); shutil.copytree(DATA_DIR,LOCAL); DATA_DIR=LOCAL
    elif os.path.isdir(LOCAL): DATA_DIR=LOCAL
except Exception as e: print("ليس Colab:",e)
print("DATA_DIR=",DATA_DIR,"| DEVICE=",DEVICE)

# ----- hyper-params (نهائية) -----
IMG_SIZE,BATCH_SIZE=224,32; LR,WEIGHT_DECAY=5e-5,1e-5
MAX_EPOCHS,PATIENCE=60,15; ALPHA,BETA=1.0,0.7
THETA_A,THETA_R,DELTA=0.75,0.60,0.15; TARGET_COVERAGE=0.70
DEFER_VAR_PERCENTILE=2.0; ACCEPT_TOP_FRAC,REVIEW_BOTTOM_FRAC,DEFER_LABEL_FRAC=0.60,0.20,0.02
N_BOOTSTRAP=1000; ACCEPT,REVIEW,DEFER=0,1,2
NORM_MEAN,NORM_STD=[0.485,0.456,0.406],[0.229,0.224,0.225]

train_tfms=transforms.Compose([transforms.Resize((IMG_SIZE,IMG_SIZE)),transforms.RandomHorizontalFlip(0.5),
    transforms.RandomRotation(10),transforms.ToTensor(),transforms.Normalize(NORM_MEAN,NORM_STD)])
eval_tfms=transforms.Compose([transforms.Resize((IMG_SIZE,IMG_SIZE)),transforms.ToTensor(),
    transforms.Normalize(NORM_MEAN,NORM_STD)])

PATIENT_RE=re.compile(r"UID[_-]?([A-Za-z]?\d+)",re.IGNORECASE)
def pid_of(p):
    m=PATIENT_RE.search(os.path.basename(p)); return m.group(1) if m else os.path.basename(p).split("_")[0]

def build_split(d):
    base=datasets.ImageFolder(d); classes,samples=base.classes,base.samples
    targets=np.array([y for _,y in samples]); groups=np.array([pid_of(p) for p,_ in samples]); idx=np.arange(len(samples))
    print(f"Classes:{classes} images:{len(samples)} patients:{len(set(groups))}")
    pat_cls={p:int(np.bincount(targets[groups==p]).argmax()) for p in np.unique(groups)}
    rng=np.random.default_rng(SEED); tr,va,te=[],[],[]
    for c in np.unique(list(pat_cls.values())):
        pats=np.array([p for p in pat_cls if pat_cls[p]==c]); rng.shuffle(pats); n=len(pats)
        ntr=max(int(round(.70*n)),1); nva=max(int(round(.15*n)),1) if n>=3 else 0
        tp,vp,ep=pats[:ntr],pats[ntr:ntr+nva],pats[ntr+nva:]
        if len(ep)==0 and n>=2: ep,vp=vp[-1:],vp[:-1]
        for p in tp: tr+=idx[groups==p].tolist()
        for p in vp: va+=idx[groups==p].tolist()
        for p in ep: te+=idx[groups==p].tolist()
    tr,va,te=map(np.array,(tr,va,te)); g=lambda ii:set(groups[ii].tolist())
    assert g(tr).isdisjoint(g(va)) and g(tr).isdisjoint(g(te)) and g(va).isdisjoint(g(te))
    for nm,ii in [("train",tr),("val",va),("test",te)]:
        b=int((targets[ii]==classes.index('blast')).sum())
        print(f"  {nm:5s}: images={len(ii):5d} patients={len(g(ii)):3d} blast={b} normal={len(ii)-b}")
    json.dump({"classes":classes,"train":tr.tolist(),"val":va.tolist(),"test":te.tolist(),
        "patients":{"train":sorted(g(tr)),"val":sorted(g(va)),"test":sorted(g(te))}},
        open(os.path.join(OUT_DIR,"patient_split.json"),"w"))
    return base,classes,samples,targets,groups,tr,va,te

def loaders(d,tr,va,te):
    return (DataLoader(Subset(datasets.ImageFolder(d,transform=train_tfms),list(tr)),BATCH_SIZE,shuffle=True,num_workers=2,pin_memory=True),
            DataLoader(Subset(datasets.ImageFolder(d,transform=eval_tfms),list(va)),BATCH_SIZE,shuffle=False,num_workers=2,pin_memory=True),
            DataLoader(Subset(datasets.ImageFolder(d,transform=eval_tfms),list(te)),BATCH_SIZE,shuffle=False,num_workers=2,pin_memory=True))

def resnet18(nc=2,dropout=0.0):
    m=models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1); f=m.fc.in_features
    m.fc=nn.Sequential(nn.Dropout(dropout),nn.Linear(f,nc)) if dropout>0 else nn.Linear(f,nc); return m.to(DEVICE)

def train_clf(model,tr,va,cw=None,tag="clf"):
    crit=nn.CrossEntropyLoss(weight=cw.to(DEVICE) if cw is not None else None)
    opt=torch.optim.Adam(model.parameters(),lr=LR,weight_decay=WEIGHT_DECAY); best,bs,bad=math.inf,None,0
    for ep in range(1,MAX_EPOCHS+1):
        model.train()
        for x,y in tr:
            x,y=x.to(DEVICE),y.to(DEVICE); opt.zero_grad(); loss=crit(model(x),y)
            if torch.isnan(loss): continue
            loss.backward(); torch.nn.utils.clip_grad_norm_(model.parameters(),1.0); opt.step()
        model.eval(); vl,n=0.0,0
        with torch.no_grad():
            for x,y in va: x,y=x.to(DEVICE),y.to(DEVICE); vl+=crit(model(x),y).item()*x.size(0); n+=x.size(0)
        vl/=max(n,1)
        if vl<best-1e-4: best,bs,bad=vl,copy.deepcopy(model.state_dict()),0
        else: bad+=1
        print(f"[{tag}] ep{ep:3d} val{vl:.4f} best{best:.4f} bad{bad}")
        if bad>=PATIENCE: break
    if bs: model.load_state_dict(bs)
    return model

@torch.no_grad()
def plogits(model,loader):
    model.eval(); L,Y=[],[]
    for x,y in loader: L.append(model(x.to(DEVICE)).cpu()); Y.append(y)
    return torch.cat(L),torch.cat(Y)

class DualHead(nn.Module):
    def __init__(s,nc=2,na=3):
        super().__init__(); bk=models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
        f=bk.fc.in_features; bk.fc=nn.Identity(); s.backbone,s.cls_head,s.dec_head=bk,nn.Linear(f,nc),nn.Linear(f,na)
    def forward(s,x): h=s.backbone(x); return s.cls_head(h),s.dec_head(h)

class IdxSub(torch.utils.data.Dataset):
    def __init__(s,d,ind,t): s.ds=datasets.ImageFolder(d,transform=t); s.ind=list(ind)
    def __len__(s): return len(s.ind)
    def __getitem__(s,i): gi=s.ind[i]; x,y=s.ds[gi]; return x,y,gi

@torch.no_grad()
def ivar(loader):
    mean=torch.tensor(NORM_MEAN).view(1,3,1,1); std=torch.tensor(NORM_STD).view(1,3,1,1)
    return np.concatenate([((x*std+mean).var(dim=(1,2,3))).numpy() for x,_ in loader])

def dec_labels(loader,baseline):
    lg,_=plogits(baseline,loader); conf=lg.softmax(1).max(1).values.numpy(); var=ivar(loader)
    dec=np.full(len(conf),-1,np.int64); sup=np.zeros(len(conf),bool)
    dm=var<=np.percentile(var,DEFER_LABEL_FRAC*100); dec[dm]=DEFER; sup[dm]=True; el=~dm
    am=el&(conf>=np.quantile(conf[el],1-ACCEPT_TOP_FRAC)); dec[am]=ACCEPT; sup[am]=True
    rm=el&~am&(conf<=np.quantile(conf[el],REVIEW_BOTTOM_FRAC)); dec[rm]=REVIEW; sup[rm]=True
    print(f"pseudo ACCEPT{am.sum()} REVIEW{rm.sum()} DEFER{dm.sum()}"); return torch.tensor(dec)

def train_dual(d,tri,vai,dl_lab,cw=None):
    tr=DataLoader(IdxSub(d,tri,train_tfms),BATCH_SIZE,shuffle=True,num_workers=2,pin_memory=True)
    va=DataLoader(IdxSub(d,vai,eval_tfms),BATCH_SIZE,shuffle=False,num_workers=2,pin_memory=True)
    model=DualHead().to(DEVICE); ccrit=nn.CrossEntropyLoss(weight=cw.to(DEVICE) if cw is not None else None)
    def dcrit(lo,dd): m=dd!=-1; return lo.sum()*0.0 if m.sum()==0 else F.cross_entropy(lo[m],dd[m])
    opt=torch.optim.Adam(model.parameters(),lr=LR,weight_decay=WEIGHT_DECAY); dl_lab=dl_lab.to(DEVICE)
    best,bs,bad=math.inf,None,0
    for ep in range(1,MAX_EPOCHS+1):
        model.train()
        for x,y,gi in tr:
            x,y=x.to(DEVICE),y.to(DEVICE); dd=dl_lab[gi.to(DEVICE)]; cl,dlg=model(x)
            loss=ALPHA*ccrit(cl,y)+BETA*dcrit(dlg,dd)
            if torch.isnan(loss) or torch.isinf(loss): continue
            opt.zero_grad(); loss.backward(); torch.nn.utils.clip_grad_norm_(model.parameters(),1.0); opt.step()
        model.eval(); vl,n=0.0,0
        with torch.no_grad():
            for x,y,gi in va:
                x,y=x.to(DEVICE),y.to(DEVICE); dd=dl_lab[gi.to(DEVICE)]; cl,dlg=model(x)
                _l=ALPHA*ccrit(cl,y)+BETA*dcrit(dlg,dd)
                if torch.isnan(_l) or torch.isinf(_l): continue
                vl+=_l.item()*x.size(0); n+=x.size(0)
        vl/=max(n,1)
        if vl<best-1e-4: best,bs,bad=vl,copy.deepcopy(model.state_dict()),0
        else: bad+=1
        print(f"[dual] ep{ep:3d} val{vl:.4f} best{best:.4f} bad{bad}")
        if bad>=PATIENCE: break
    model.load_state_dict(bs); torch.save(model.state_dict(),os.path.join(OUT_DIR,"dual_head_resnet18.pt")); return model

def policy(dp,var,vt,ta=THETA_A,tr_=THETA_R,dl=DELTA):
    pa,pr=dp[:,ACCEPT],dp[:,REVIEW]; dec=np.full(len(dp),REVIEW)
    dec[(pa>ta)&(pa>pr+dl)]=ACCEPT; dec[var<=vt]=DEFER; return dec

def smetrics(y,pb,dec,bi=0):
    am=dec==ACCEPT; o=dict(coverage=float(am.mean()),review_rate=float((dec==REVIEW).mean()),
        defer_rate=float((dec==DEFER).mean()),accepted_n=int(am.sum()))
    if am.sum()==0: o.update(acc_accuracy=np.nan,auc=np.nan,blast_sensitivity=np.nan,blast_fnr=np.nan); return o
    yt,p=y[am],pb[am]; yp=(p>=0.5).astype(int); yb=(yt==bi).astype(int)
    o["acc_accuracy"]=float(accuracy_score(yb,yp)); o["auc"]=float(roc_auc_score(yb,p)) if len(np.unique(yb))>1 else np.nan
    tp=int(((yp==1)&(yb==1)).sum()); fn=int(((yp==0)&(yb==1)).sum())
    o["blast_sensitivity"]=tp/(tp+fn+1e-9); o["blast_fnr"]=fn/(tp+fn+1e-9); return o

def boot(y,pb,dec,pid,bi=0):
    pid=np.asarray(pid); uq=np.unique(pid); rng=np.random.default_rng(SEED); s,f=[],[]
    for _ in range(N_BOOTSTRAP):
        ch=rng.choice(uq,len(uq),replace=True); mk=np.concatenate([np.where(pid==p)[0] for p in ch])
        m=smetrics(y[mk],pb[mk],dec[mk],bi)
        if not np.isnan(m["blast_sensitivity"]): s.append(m["blast_sensitivity"]); f.append(m["blast_fnr"])
    ci=lambda a:[float(np.percentile(a,2.5)),float(np.percentile(a,97.5))]
    return {"sensitivity_CI":ci(s),"fnr_CI":ci(f)}

def ece_fn(probs,labels,nb=10):
    cf,pr=probs.max(1),probs.argmax(1); ac=(pr==labels).astype(float); bins=np.linspace(0,1,nb+1); e=0.0; tab=[]
    for i in range(nb):
        m=(cf>bins[i])&(cf<=bins[i+1])
        if m.sum()==0: tab.append((bins[i],bins[i+1],0,np.nan,np.nan)); continue
        a,c=ac[m].mean(),cf[m].mean(); e+=(m.sum()/len(cf))*abs(a-c); tab.append((bins[i],bins[i+1],int(m.sum()),float(a),float(c)))
    return e,tab

@torch.no_grad()
def mcdrop(model,loader,passes=20):
    for m in model.modules():
        if isinstance(m,nn.Dropout): m.train()
    P=[]
    for _ in range(passes): P.append(torch.cat([model(x.to(DEVICE)).softmax(1).cpu() for x,_ in loader]))
    P=torch.stack(P); mp=P.mean(0).numpy(); return mp,-(mp*np.log(mp+1e-12)).sum(1)

def sel_unc(y,pb,unc,pid,cap=0.35,bi=0):
    best=None
    for q in np.linspace(0.05,0.95,19):
        thr=np.quantile(unc,q); dec=np.where(unc<=thr,ACCEPT,REVIEW); m=smetrics(y,pb,dec,bi)
        if m["review_rate"]<=cap and not np.isnan(m["blast_fnr"]):
            if best is None or m["blast_fnr"]<best[0]: best=(m["blast_fnr"],m,dec)
    if best is None:
        thr=np.quantile(unc,0.65); dec=np.where(unc<=thr,ACCEPT,REVIEW); best=(np.nan,smetrics(y,pb,dec,bi),dec)
    _,m,dec=best; m.update(boot(y,pb,dec,pid,bi)); return m

class GradCAM:
    def __init__(s,model,layer):
        s.model=model; s.a=s.g=None
        layer.register_forward_hook(lambda m,i,o:setattr(s,"a",o.detach()))
        layer.register_full_backward_hook(lambda m,gi,go:setattr(s,"g",go[0].detach()))
    def __call__(s,x,head="dec",target=ACCEPT):
        s.model.zero_grad(); cl,dl=s.model(x); (dl if head=="dec" else cl)[:,target].sum().backward()
        w=s.g.mean(dim=(2,3),keepdim=True); cam=F.relu((w*s.a).sum(1)); cam=cam/(cam.amax(dim=(1,2),keepdim=True)+1e-8); return cam.cpu().numpy()

def cmask(xu): return (xu.mean(0)>0.05).cpu().numpy()
def ametrics(cam,mask):
    eps=1e-8; p=cam/(cam.sum()+eps); ent=float(-(p*np.log(p+eps)).sum()); inside=cam[mask]
    pk=float(cam.max()/(cam.mean()+1e-6)); fr=float(inside.sum()/(cam.sum()+eps))
    H,W=cam.shape; ys,xs=np.mgrid[0:H,0:W]; cc=np.array([(cam*ys).sum(),(cam*xs).sum()])/(cam.sum()+eps)
    cm=mask.astype(float); mc=np.array([(cm*ys).sum(),(cm*xs).sum()])/(cm.sum()+eps)
    return dict(entropy=ent,peakiness=pk,focus_ratio=fr,com_distance=float(np.linalg.norm(cc-mc)/math.hypot(H,W)))
def cohend(a,b):
    a,b=np.asarray(a),np.asarray(b); na,nb=len(a),len(b)
    sp=math.sqrt(((na-1)*a.var(ddof=1)+(nb-1)*b.var(ddof=1))/(na+nb-2)+1e-12); return float((a.mean()-b.mean())/(sp+1e-12))
def cmp_att(acc,rev,k=4):
    r={}
    for key in acc:
        u,p=mannwhitneyu(acc[key],rev[key],alternative="two-sided")
        r[key]=dict(p_value=float(p),p_bonferroni=float(min(p*k,1.0)),cohens_d=cohend(acc[key],rev[key]),
            accept_mean=float(np.mean(acc[key])),review_mean=float(np.mean(rev[key])))
    return r

# ---------- الأشكال ----------
def _box(ax,x,y,w,h,t,fc):
    ax.add_patch(FancyBboxPatch((x,y),w,h,boxstyle="round,pad=0.02",lw=1.5,edgecolor="#333",facecolor=fc))
    ax.text(x+w/2,y+h/2,t,ha="center",va="center",fontsize=11,weight="bold")
def _arr(ax,x1,y1,x2,y2):
    ax.add_patch(FancyArrowPatch((x1,y1),(x2,y2),arrowstyle="-|>",mutation_scale=18,lw=1.5,color="#333"))

def fig1(samples,targets,classes,n=4):
    bi,ni=classes.index('blast'),classes.index('normal')
    bx=np.where(targets==bi)[0][:n]; nx=np.where(targets==ni)[0][:n]
    fig,ax=plt.subplots(2,n,figsize=(2.4*n,5))
    for j,i in enumerate(bx): ax[0,j].imshow(Image.open(samples[i][0])); ax[0,j].axis('off')
    for j,i in enumerate(nx): ax[1,j].imshow(Image.open(samples[i][0])); ax[1,j].axis('off')
    ax[0,0].set_title("Blast",fontsize=12); ax[1,0].set_title("Normal",fontsize=12)
    fig.suptitle("Figure 1. Representative C-NMC images",fontsize=13)
    plt.tight_layout(); plt.savefig(os.path.join(FIG_DIR,"fig1_examples.png"),dpi=300,bbox_inches='tight'); plt.close(); print("fig1")

def fig2():
    fig,ax=plt.subplots(figsize=(11,4.2)); ax.set_xlim(0,11); ax.set_ylim(0,4.2); ax.axis("off")
    _box(ax,0.2,1.6,1.7,1.0,"Input\n224x224x3","#E8F0FE"); _box(ax,2.4,1.6,2.2,1.0,"ResNet-18\nbackbone\n(shared)","#FCE8E6")
    _box(ax,5.3,2.7,2.4,1.0,"Classification head\n(blast/normal)","#E6F4EA"); _box(ax,5.3,0.5,2.4,1.0,"Decision head\n(ACCEPT/REVIEW/DEFER)","#FEF7E0")
    _box(ax,8.4,2.7,2.3,1.0,"p(blast)","#E6F4EA"); _box(ax,8.4,0.5,2.3,1.0,"Decision policy\nθa,θr,δ","#FEF7E0")
    _arr(ax,1.9,2.1,2.4,2.1); _arr(ax,4.6,2.1,5.3,3.2); _arr(ax,4.6,2.1,5.3,1.0); _arr(ax,7.7,3.2,8.4,3.2); _arr(ax,7.7,1.0,8.4,1.0)
    ax.set_title("Figure 2. Dual-head architecture",fontsize=13,weight="bold")
    plt.tight_layout(); plt.savefig(os.path.join(FIG_DIR,"fig2_architecture.png"),dpi=300,bbox_inches="tight"); plt.close(); print("fig2")

def fig3():
    fig,ax=plt.subplots(figsize=(12,3.2)); ax.set_xlim(0,12); ax.set_ylim(0,3.2); ax.axis("off")
    steps=[("C-NMC\ndata","#E8F0FE"),("Patient\nsplit","#E8F0FE"),("Baseline\nclassifier","#FCE8E6"),
        ("Pseudo-\nlabels","#FEF7E0"),("Dual-head\ntraining","#FCE8E6"),("Selective\neval+ECE","#E6F4EA"),("Grad-CAM\n+attention","#E6F4EA")]
    x=0.2; w=1.5
    for i,(t,c) in enumerate(steps):
        _box(ax,x,1.0,w,1.2,t,c)
        if i<len(steps)-1: _arr(ax,x+w,1.6,x+w+0.18,1.6)
        x+=w+0.18
    ax.set_title("Figure 3. Training and inference pipeline",fontsize=13,weight="bold")
    plt.tight_layout(); plt.savefig(os.path.join(FIG_DIR,"fig3_pipeline.png"),dpi=300,bbox_inches="tight"); plt.close(); print("fig3")

def fig4(model,loader,cam,vt,n=3):
    mt=torch.tensor(NORM_MEAN).view(3,1,1); st=torch.tensor(NORM_STD).view(3,1,1); shown=0
    fig,ax=plt.subplots(2,n,figsize=(2.6*n,5.2))
    for x,y in loader:
        dp=model(x.to(DEVICE))[1].softmax(1).detach().cpu().numpy(); dec=policy(dp,ivar([(x,y)]),vt)
        c=cam(x.to(DEVICE),head="dec",target=ACCEPT)
        crs=F.interpolate(torch.tensor(c)[:,None],size=(IMG_SIZE,IMG_SIZE),mode="bilinear",align_corners=False)[:,0].numpy()
        for i in range(x.size(0)):
            if shown>=n: break
            raw=(x[i]*st+mt).permute(1,2,0).numpy().clip(0,1)
            ax[0,shown].imshow(raw); ax[0,shown].axis('off'); ax[0,shown].set_title(["ACCEPT","REVIEW","DEFER"][dec[i]],fontsize=11)
            ax[1,shown].imshow(raw); ax[1,shown].imshow(crs[i],cmap='jet',alpha=0.5); ax[1,shown].axis('off'); shown+=1
        if shown>=n: break
    fig.suptitle("Figure 4. Decision-level Grad-CAM",fontsize=12)
    plt.tight_layout(); plt.savefig(os.path.join(FIG_DIR,"fig4_gradcam.png"),dpi=300,bbox_inches="tight"); plt.close(); print("fig4")

def fig5(tab,ece):
    mids=[(a+b)/2 for a,b,n,ac,cf in tab]; accs=[ac for *_,ac,_ in tab]
    fig,ax=plt.subplots(figsize=(5,5)); ax.plot([0,1],[0,1],'--',color='gray',label='Perfect')
    ax.bar(mids,[a if not np.isnan(a) else 0 for a in accs],width=0.09,alpha=0.8,edgecolor='black',label='Model')
    ax.set_xlabel('Confidence',fontsize=12); ax.set_ylabel('Accuracy',fontsize=12)
    ax.set_title(f'Figure 5. Reliability (ECE={ece:.3f})',fontsize=12); ax.legend(fontsize=10)
    plt.tight_layout(); plt.savefig(os.path.join(FIG_DIR,"fig5_reliability.png"),dpi=300,bbox_inches="tight"); plt.close(); print("fig5")

# ============ التشغيل ============
base,classes,samples,targets,groups,tri,vai,tei=build_split(DATA_DIR)
bi=classes.index("blast")
cnt=np.bincount(targets[tri],minlength=len(classes)).astype(float)
cw=torch.tensor(cnt.sum()/(len(classes)*(cnt+1e-9)),dtype=torch.float32)
trL,vaL,teL=loaders(DATA_DIR,tri,vai,tei)

print("\n==Baseline=="); baseline=train_clf(resnet18(),trL,vaL,cw,"baseline")
print("\n==Pseudo-labels==")
dfull=torch.full((len(samples),),-1,dtype=torch.long)
trE=DataLoader(Subset(datasets.ImageFolder(DATA_DIR,transform=eval_tfms),list(tri)),BATCH_SIZE,shuffle=False,num_workers=2)
dtr=dec_labels(trE,baseline)
for loc,gi in enumerate(tri): dfull[gi]=dtr[loc]
print("\n==Dual-head=="); model=train_dual(DATA_DIR,tri,vai,dfull,cw)

print("\n==Test=="); vt=np.percentile(ivar(trE),DEFER_VAR_PERCENTILE)
model.eval(); cpb,dpr,ya=[],[],[]; tpaths=[samples[i][0] for i in tei]
with torch.no_grad():
    for x,y in teL:
        cl,dl=model(x.to(DEVICE)); cpb.append(cl.softmax(1)[:,bi].cpu().numpy()); dpr.append(dl.softmax(1).cpu().numpy()); ya.append(y.numpy())
cpb=np.concatenate(cpb); dpr=np.concatenate(dpr); ya=np.concatenate(ya); tv=ivar(teL); pid=np.array([pid_of(p) for p in tpaths])
bt,bg=THETA_A,1e9
for t in np.linspace(0.30,0.90,25):
    cov=(policy(dpr,tv,vt,ta=t)==ACCEPT).mean()
    if abs(cov-TARGET_COVERAGE)<bg: bg,bt=abs(cov-TARGET_COVERAGE),t
print("matched theta_a=",round(bt,3)); decision=policy(dpr,tv,vt,ta=bt)
m=smetrics(ya,cpb,decision,bi); m.update(boot(ya,cpb,decision,pid,bi)); print("PROPOSED:",json.dumps(m,indent=2,default=float))

print("\n==Calibration=="); tl,ty=plogits(baseline,teL); probs=tl.softmax(1).numpy(); ece,tab=ece_fn(probs,ty.numpy()); print("ECE=",round(ece,4))

print("\n==Baselines=="); t3={"Proposed":m}
bl,_=plogits(baseline,teL); bp=bl.softmax(1).numpy()
t3["ConfidenceThresholding"]=sel_unc(ya,bp[:,bi],1-bp.max(1),pid,bi=bi)
mc=train_clf(resnet18(dropout=0.5),trL,vaL,cw,"mcdrop"); mp,mu=mcdrop(mc,teL,20); t3["MC_Dropout"]=sel_unc(ya,mp[:,bi],mu,pid,bi=bi)
ep=[]
for s in range(3):
    torch.manual_seed(SEED+s); em=train_clf(resnet18(),trL,vaL,cw,f"ens{s}"); ep.append(plogits(em,teL)[0].softmax(1).numpy())
ens=np.stack(ep); t3["DeepEnsemble"]=sel_unc(ya,ens.mean(0)[:,bi],ens.std(0).mean(1),pid,bi=bi)
json.dump(t3,open(os.path.join(OUT_DIR,"table3_comparison.json"),"w"),indent=2,default=float)

print("\n==Attention=="); cam=GradCAM(model,model.backbone.layer4[-1])
acc={k:[] for k in ["entropy","peakiness","focus_ratio","com_distance"]}; rev={k:[] for k in acc}
mt=torch.tensor(NORM_MEAN).view(3,1,1); st=torch.tensor(NORM_STD).view(3,1,1); seen=0
for x,y in teL:
    dp=model(x.to(DEVICE))[1].softmax(1).detach().cpu().numpy(); db=policy(dp,ivar([(x,y)]),vt)
    c=cam(x.to(DEVICE),head="dec",target=ACCEPT)
    crs=F.interpolate(torch.tensor(c)[:,None],size=(IMG_SIZE,IMG_SIZE),mode="bilinear",align_corners=False)[:,0].numpy()
    for i in range(x.size(0)):
        a=ametrics(crs[i],cmask(x[i]*st+mt)); tg=acc if db[i]==ACCEPT else (rev if db[i]==REVIEW else None)
        if tg is not None:
            for k in a: tg[k].append(a[k])
    seen+=x.size(0)
    if seen>=1000: break
att=cmp_att(acc,rev); json.dump({"accept_n":len(acc["entropy"]),"review_n":len(rev["entropy"]),"tests":att},
    open(os.path.join(OUT_DIR,"attention_metrics.json"),"w"),indent=2,default=float); print(json.dumps(att,indent=2,default=float))

print("\n==Figures=="); fig1(samples,targets,classes); fig2(); fig3(); fig4(model,teL,cam,vt); fig5(tab,ece)
json.dump({"proposed_test":m,"ECE":ece,"attention_tests":att,
    "split":{"patients":len(set(groups)),"train":len(tri),"val":len(vai),"test":len(tei)}},
    open(os.path.join(OUT_DIR,"RESULTS_SUMMARY.json"),"w"),indent=2,default=float)
print("\n>>> تم. النتائج والأشكال في",OUT_DIR)
```
